```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-


import random
import re
import nltk
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
# Import LabelEncoder
from sklearn import preprocessing
# import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
nltk.download('swadesh')

from nltk.corpus import swadesh
en2es = swadesh.entries(['en', 'es'])
translation = dict(en2es)

#weather label variables
weather = ['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny', 'Rainy','Sunny','Overcast','Overcast','Rainy']

# output class
play = ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']

# creating LabelEncoder
labelCode = preprocessing.LabelEncoder()

# Converting string labels into numbers.
weather_encoded=labelCode.fit_transform(play)

# import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

# create a Gaussian Classifier
model = GaussianNB()

# converting string labels into numbers.
label=labelCode.fit_transform(play)

# converting 1D array to 2D
weather_2d = np.reshape(weather_encoded, (-1, 1))

# train the model using the training sets
model.fit(weather_2d, label)

# predicting the odel
predicted= model.predict([[0]])

# Ask the user for input


# Sentences we'll respond with if the user greeted us
GREETING_KEYWORDS = ("hello", "hi", "greetings", "sup", "what's up", "wassup", "howdy", "Hello", "Hi", "Greetings", "Sup", "What's up", "Wassup", "Howdy")

GREETING_RESPONSES = ["sup", "suh", "*nods*", "k", "howdy", "Sup", "Suh", "K", "Howdy"]

JOKES = ["Did you hear about the first restaurant to open on the moon? \n It had great food, but no atmosphere.",
                  "Do you want to hear a construction joke? \n Sorry, I’m still working on it.", 
                    "When does a joke become a ‘dad’ joke? \n When it becomes apparent."]


#If any of the words in the user's input was a greeting, return a greeting response
def check_for_greeting(sentence):
    for word in sentence.split(" "):
        if word.lower() in GREETING_KEYWORDS:
            #Photo = cv2.imread('ron.jpg')
            #plt.imshow(Photo)
            #plt.show()
            return random.choice(GREETING_RESPONSES)
  
# Example: check for a word "dog" and respond to it.  
def check_for_phrase(sentence):
    KEYPHRASES = ["dog", "Dog"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            #dogPhoto = cv2.imread('dog.jpeg') photos are preventing request for
            #cv2_imshow(dogPhoto)              input on some entries.
            return "This is my response to dog."

#translation example
def check_for_translation(sentence):
    KEYPHRASES = ["translate", "Translate"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            b =  input('Type the word you want to translate from English to Spanish: ')
            output = translation[b]
            return output
          
# Example: check for a word "cat" and respond to it.  
def check_for_cat(sentence):
    KEYPHRASES = ["cat", "Cat"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            #catPhoto = cv2.imread('cat.jpeg')
            #cv2_imshow(catPhoto)
            return "This is my response to cat."

# Example: check for a word "image" and respond to it.  
def check_for_image(sentence):
   KEYPHRASES = ["image", "img", "photo", "Photo", "Image", "Img"]
   for word in sentence.split(" "):
        if word in KEYPHRASES:
            from google.colab import files
            print("Choose the image you want to upload.")
            uploaded = files.upload()
            i = uploaded.keys()
            up = cv2.imread(list(i)[0])
            cv2_imshow(up)
            return 'This is the image you uploaded.'

#Cropping Image
def crop_image(sentence):
    KEYPHRASES = ["Crop", "crop", "zoom", "Zoom"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            j = input('Type the name of the image file you want to crop: ')
            k = cv2.imread(j)
            height = k.shape[0]
            width = k.shape[1]
            newA = height // 10
            newB = width // 10
            crop = k[height - 100:height - newA, width - 100:width - newB]
            cv2_imshow(crop)
            #print(k.shape)
            return "Here is your image cropped to my liking."

#weather naive bayes
def check_for_weather(sentence):
    KEYPHRASES = ["weather", "Weather"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            return print("1 is good weather, 0 is bad weather", "\n Your weather score is: ", predicted)

# Example: get your chatbot to tell a story. The \n command
# starts a new line (useful to make the ouptut readable.)
def tellstory(sentence):
    KEYPHRASES = ["story", "Story"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            s = "Here is my story \n It was a sunny day at the park, and we had a frisbee \n until a dog that was passing by took it away. \n Now we no longer have a frisbee."
            return s
          
#jokes
def jokes(sentence):
    KEYPHRASES = ["joke", "Joke", "Jokes", "jokes"]
    for word in sentence.split(" "):
        if word.lower() in KEYPHRASES:
            return random.choice(JOKES)

# Reflections swap the users pronouns back at them. For instance, if the user
# says: "I need you". The response will flip "I -> you", "you -> me", so the 
# bot will say "you need me". 
reflections = {
    "am": "are",
    "was": "were",
    "i": "you",
    "i'd": "you would",
    "i've": "you have",
    "i'll": "you will",
    "my": "your",
    "are": "am",
    "you've": "I have",
    "you'll": "I will",
    "your": "my",
    "yours": "mine",
    "you": "me",
    "me": "you"
}

# This function looks for the starting phrase, and then takes the rest of the 
# user's sentence after the phrase and puts it in variable 0
psychobabble = [
    [r'I need (.*)',
     ["What is {0}?",
      "How do you get {0}?",
      "Really? {0}?"]],
    
    [r'Who are you?',
     ["I'm too advanced for your tiny brain to comprehend."]], 

    [r'who are you?',
     ["I'm too advanced for your tiny brain to comprehend."]],
    
    [r'I like (.*)',
     ["Bruh, really?"]],
 
    [r'Why don\'?t you ([^\?]*)\??',
     ["Why would I ever {0}?",
      "I'll never {0}.",
      "You probably should {0}."]],
      
    [r'I want (.*)',
     ["TEST"]],
    # This must be the last block
    [r'(.*)',
     ["Please don't tell me more.",
      "Let's change focus a bit... Ask anything about me.",
      "Can you say something else?",
      "Really, {0}?",
      "Bruh.",
      "Cool.",
      "{0}.",
      "What do you think?",
      "You need actual help.",
      "Seriously, you need help."]],
      
]
        
     
def reflect(fragment):
    tokens = fragment.lower().split()
    for i, token in enumerate(tokens):
        if token in reflections:
            tokens[i] = reflections[token]
    return ' '.join(tokens)
 
def analyze(statement):
    for pattern, responses in psychobabble:
        match = re.match(pattern, statement.rstrip(".!"))
        if match:
            response = random.choice(responses)
            return response.format(*[reflect(g) for g in match.groups()])


## This is the main part of the Chatbot code, checks for phrases, words, and 
## prints responses
```
